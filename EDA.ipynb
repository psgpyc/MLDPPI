```python
import matplotlib.pyplot as plt

# Example categorical columns and custom tick mappings
categorical_cols = ['smoking', 'high_bp', 'high_chol', 'heart_diseases',
                    'asthma', 'kidney_diseases', 'marital_status',
                    'education', 'general_health', 'arthritis',
                    'depression', 'sex', 'race']

# Custom tick labels for each column (example mapping)
custom_ticks = {
    'smoking': ["Never Smoked", "Former Smoker", "Smokes Everyday", "Smokes Someday", "Refused"],
    'high_bp': ['Yes', 'No', 'Dont Know'],
    'high_chol': ['Yes', 'No', 'Dont Know'],
    'heart_diseases': ['No Heart Disease', 'Heart Disease'],
    'asthma': ['No Asthma', 'Asthma', 'Had Asthma', 'Dont Know'],
    'kidney_diseases': ['No Kidney Disease', 'Kidney Disease', 'Dont Know', 'Refused To Answer'],
    'marital_status': ['Married', 'Widowed' ,'Divorced', 'Never Married', 'Living Together', 'Seperated'  ,'Refused To Answer'],
    'education': ['High School', 'Some High School', 'Elementary', 'Never Attended', 'Refused To Answer'],
    'general_health': ['Excellent', 'Good', 'Refused To Answer'],
    'arthritis': ['No Arthritis', 'Arthritis'],
    'depression': ['No Depression', 'Depression'],
    'sex': ['Female', 'Male'],
    'race': ['White', 'Others', 'Black/African American', 'Asian', 'Mixed', 'Hispanic/Latino', 'Refused To Answer', 'American Indian', 'Pacific Islander']
}

# Iterate through each column
for col in categorical_cols:
    ax = df[col].value_counts(normalize=True).plot(kind='bar', title=col, color='skyblue', edgecolor='black')

    # Add custom x-ticks if available in the dictionary
    if col in custom_ticks:
        tick_positions = range(len(custom_ticks[col]))
        plt.xticks(ticks=tick_positions, labels=custom_ticks[col], rotation=45, fontsize=10)

    # Add axis labels
    plt.ylabel('Proportion', fontsize=12)

    # Add grid for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show plot
    plt.tight_layout()  # Adjust layout to prevent overlap
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Example categorical columns and custom tick mappings
categorical_cols = ['smoking', 'high_bp', 'high_chol', 'heart_diseases',
                    'asthma', 'kidney_diseases', 'marital_status',
                    'education', 'general_health', 'arthritis',
                    'depression', 'sex', 'race']

# Custom tick labels for each column (example mapping)
custom_ticks = {
    'smoking': ["Never Smoked", "Former Smoker", "Smokes Everyday", "Smokes Someday", "Refused"],
    'high_bp': ['Yes', 'No', 'Dont Know'],
    'high_chol': ['Yes', 'No', 'Dont Know'],
    'heart_diseases': ['No Heart Disease', 'Heart Disease'],
    'asthma': ['No Asthma', 'Asthma', 'Had Asthma', 'Dont Know'],
    'kidney_diseases': ['No Kidney Disease', 'Kidney Disease', 'Dont Know', 'Refused To Answer'],
    'marital_status': ['Married', 'Widowed' ,'Divorced', 'Never Married', 'Living Together', 'Seperated'  ,'Refused To Answer'],
    'education': ['High School', 'Some High School', 'Elementary', 'Never Attended', 'Refused To Answer'],
    'general_health': ['Excellent', 'Good', 'Refused To Answer'],
    'arthritis': ['No Arthritis', 'Arthritis'],
    'depression': ['No Depression', 'Depression'],
    'sex': ['Female', 'Male'],
    'race': ['White', 'Others', 'Black/African American', 'Asian', 'Mixed', 'Hispanic/Latino', 'Refused To Answer', 'American Indian', 'Pacific Islander']
}

# Iterate through each column
for col in categorical_cols:
    ax = pd.crosstab(df[col], df['has_diabetes'], normalize='index').plot(kind='bar', title=col, edgecolor='black')

    # Add custom x-ticks if available in the dictionary
    if col in custom_ticks:
        tick_positions = range(len(custom_ticks[col]))
        plt.xticks(ticks=tick_positions, labels=custom_ticks[col], rotation=45, fontsize=10)

    # Add axis labels
    plt.ylabel('Proportion', fontsize=12)

    # Add grid for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show plot
    plt.tight_layout()  # Adjust layout to prevent overlap
    plt.show()
```
